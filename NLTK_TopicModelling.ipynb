```
from gensim import corpora, models, similarities
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.stem import WordNetLemmatizer
from nltk.tag import pos_tag
import nltk
from nltk.corpus import stopwords
from operator import itemgetter
import re
import csv
import pprint

nltk.download('stopwords')
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('wordnet')
```

```
! pip install wikipedia
import wikipedia

wikipedia = wikipedia.page("Coronnavirus").content
print(wikipedia)
```

```
# Sentence Tokenization

sents = sent_tokenize(wikipedia)
print(sents)
```

```
# Word Tokenization
words = []
for sent in sents:
  words.extend(word_tokenize(sent))
print(words)
```

```
# POS Tagging
tagged = pos_tag(words)
print(tagged)
```

```
from nltk.corpus import wordnet

def getpos(tag):
  if tag.startswith("J"):
    return wordnet.ADJ
  elif tag.startswith("V"):
    return wordnet.VERB
  elif tag.startswith("N"):
    return wordnet.NOUN
  elif tag.startswith("R"):
    return wordnet.ADV
  else:
    return ""
  
```

```
# Lemmatization woth POS Tags
punctuation = u",.?!@#$%^&*()-_<>':;{}[]\"\'\n\\\t\r|+=/"
stop_words = set(stopwords.words("english"))
wordnet_lemmatizer = WordNetLemmatizer()
lemma_pos = []
for word, tag in tagged:
  if word not in stop_words and word not in punctuation and word.isalpha():
    p = getpos(tag)
    if p != '':
      l = wordnet_lemmatizer.lemmatize(word, pos = p)
      lemma_pos.append(l)

print(lemma_pos)
```

```
# Convert words to number
id2word = corpora.Dictionary([lemma_pos])
texts = lemma_pos
# Bag of Words
corpus = [id2word.doc2bow([text]) for text in texts]

n_topics = 5
lda_model = models.LdaModel(
    corpus=corpus,
    id2word=id2word,
    num_topics=n_topics,
    random_state = 100,
    update_every = 1,
    chunksize = 100,
    passes = 10,
    alpha = "symmetric",
    per_word_topics = True
)

print(lda_model.print_topics())
```

```
from collections import Counter
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors

topics = lda_model.show_topics(formatted=False)
data_flat = [w for w_list in lemma_pos for w in w_list]
counter = Counter(data_flat)

out = []
for i, topic in topics:
  for word, weight in topic:
    out.append([word, i, weight, counter[word]])

df = pd.DataFrame(out, columns = ['word', 'topic_id', 'importance', 'word_count'])

df
```
