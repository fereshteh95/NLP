```
from gensim import corpora, models, similarities
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.stem import WordNetLemmatizer
from nltk.tag import pos_tag
import nltk
from nltk.corpus import stopwords
from operator import itemgetter
import re
import csv
import pprint

nltk.download('stopwords')
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('wordnet')
```

```
! pip install wikipedia
import wikipedia

wikipedia = wikipedia.page("Artificial_intelligence").content
print(wikipedia)
```

```
# Sentence Tokenization

sents = sent_tokenize(wikipedia)
print(sents)
```

```
# Word Tokenization
words = []
for sent in sents:
  words.extend(word_tokenize(sent))
print(words)
```

```
# POS Tagging
tagged = pos_tag(words)
print(tagged)
```

```
from nltk.corpus import wordnet

def getpos(tag):
  if tag.startswith("J"):
    return wordnet.ADJ
  elif tag.startswith("V"):
    return wordnet.VERB
  elif tag.startswith("N"):
    return wordnet.NOUN
  elif tag.startswith("R"):
    return wordnet.ADV
  else:
    return ""
  
```

```
# Lemmatization woth POS Tags
punctuation = u",.?!@#$%^&*()-_<>':;{}[]\"\'\n\\\t\r|+=/"
stop_words = set(stopwords.words("english"))
wordnet_lemmatizer = WordNetLemmatizer()
lemma_pos = []
for word, tag in tagged:
  if word not in stop_words and word not in punctuation and word.isalpha():
    p = getpos(tag)
    if p != '':
      l = wordnet_lemmatizer.lemmatize(word, pos = p)
      lemma_pos.append(l)

print(lemma_pos)
```

```
word_frequencies = {}
for word in words:
  if word not in stop_words and word.isalpha():
    if word not in word_frequencies.keys():
      word_frequencies[word] = 1
    else:
      word_frequencies[word] += 1

print(word_frequencies)
```

```
maximum_frequency = max(word_frequencies.values())

for word in word_frequencies.keys():
  word_frequencies[word] = (word_frequencies[word]/maximum_frequency)

print(word_frequencies)
```

```
sentence_scores = {}

for sent in sents:
  for word in words:
    if word.lower() in word_frequencies.keys():
      if len(sent.split(' ')) < 30:
        if sent not in sentence_scores.keys():
          sentence_scores[sent] = word_frequencies[word]
        else:
          sentence_scores[sent] += word_frequencies[word]


print(sentence_scores)
```

```
import heapq
summary_sentences = heapq.nlargest(7, sentence_scores, key=sentence_scores.get)

summary = " ".join(summary_sentences)
print(summary)
```
